```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Data.csv')  

# Filter the DataFrame for rows where LLM is "GPT 4o"
GPT4oData = df[df['Model'] == 'GPT 4']

plt.figure(figsize=(12, 6))
sns.barplot(data=GPT4oData, x='Question', y='Semantic Entropy', color='skyblue')

plt.xticks(rotation=45, ha='right')
plt.title('Semantic Entropy by Question (GPT 4o)')
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for rows where LLM is "GPT 4o"
Deepseek = df[df['Model'] == 'Deepseek V3']

plt.figure(figsize=(12, 6))
sns.barplot(data=Deepseek, x='Question', y='Semantic Entropy', color='skyblue')

plt.xticks(rotation=45, ha='right')
plt.title('Semantic Entropy by Question (DeepSeek-V3)')
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for rows where LLM is Claude
Deepseek = df[df['Model'] == 'Claude']

plt.figure(figsize=(12, 6))
sns.barplot(data=Deepseek, x='Question', y='Semantic Entropy', color='skyblue')

plt.xticks(rotation=45, ha='right')
plt.title('Semantic Entropy by Question (Claude 3.7 Sonnet)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(
    data=df,
    x='Question',
    y='Semantic Entropy',
    hue='Model',
    ci=None,               # turn off confidence intervals if you only have single values
    dodge=True             # separate bars side by side
)


plt.xticks(rotation=45, ha='right')
plt.ylabel('Semantic Entropy')
plt.title('Semantic Entropy by Question and Model')
plt.legend(title='Model', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
semantic_entropy_sum = df.groupby("Model")["Semantic Entropy"].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=semantic_entropy_sum, x='Model', y='Semantic Entropy', color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Total Semantic Entropy by Model')
plt.tight_layout()
plt.show()
```
